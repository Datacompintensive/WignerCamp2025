## Introduction to `Matplotlib`
`Matplotlib` is a plotting library for the `Python` programming language and its numerical mathematics extension `NumPy`. It provides an object-oriented `API` for embedding plots into applications.

```python
# Importing Matplotlib
import matplotlib.pyplot as plt
```

## Changing Markers, Line Styles, and Colors
You can customize the appearance of your plots by changing markers, line styles, and colors.

```python
# Example plot with different markers, line styles, and colors
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker='o', linestyle='-', color='r', label='Line 1')
plt.plot(x, [15, 25, 20, 35], marker='s', linestyle='--', color='b', label='Line 2')
plt.legend()
```

## Logarithmic Scales
You can set the scales of the `x`-axis and `y`-axis to logarithmic scales.

```python
# Example plot with logarithmic scales
x = [1, 10, 100, 1000]
y = [1, 10, 100, 1000]

plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
```

## Aspect Ratios
You can control the aspect ratio of your plots using the `set_aspect` method.

```python
# Example plot with different aspect ratios
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.gca().set_aspect('equal')
```

## Adding Text to Plots
You can add text annotations to your plots using the `text` function.

```python
# Example plot with text annotations
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker='o')
plt.text(2, 20, 'This is a point')
```

### Task 4

## Titles
You can add titles to your plots using the `title` function.

```python
# Example plot with a title
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, marker='o')
plt.title('Sample Plot')
```

## Subplots
You can create multiple subplots in a single figure using the `subplots` function.

```python
# Example with multiple subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# First subplot
x1 = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
ax[0].plot(x1, y1, marker='o')
ax[0].set_title('Subplot 1')

# Second subplot
x2 = [1, 2, 3, 4]
y2 = [30, 25, 20, 10]
ax[1].plot(x2, y2, marker='s')
ax[1].set_title('Subplot 2')
```

### Task 6
Create a figure with 2x1 subplots:
1. First subplot: `x = [1, 2, 3, 4]`, `y = [5, 10, 15, 20]`, title 'First Plot'
2. Second subplot: `x = [1, 2, 3, 4]`, `y = [20, 15, 10, 5]`, title 'Second Plot'

```python
# Task 6 Solution
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# First subplot
x1 = [1, 2, 3, 4]
y1 = [5, 10, 15, 20]
ax[0].plot(x1, y1, marker='o')
ax[0].set_title('First Plot')

# Second subplot
x2 = [1, 2, 3, 4]
y2 = [20, 15, 10, 5]
ax[1].plot(x2, y2, marker='s')
ax[1].set_title('Second Plot')
```
